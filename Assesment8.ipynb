```python
def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    else:
        return "Fail"

print(calculate_grade(80))
```

```python
#2.Even or Odd number checker
```

```python
def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False
        
print(is_even(4))   
print(is_even(7))   
```
