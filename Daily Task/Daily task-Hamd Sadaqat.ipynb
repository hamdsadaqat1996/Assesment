```python
setA = {3, 4, 5, 6}
```

```python
setB = {5, 6, 7, 8}
```

```python
setA.add(10)
```

```python
print("After add:", setA)
```

```python
union_set = setA.union(setB)
```

```python
print("Union:", union_set)
```

```python
#intersection
```

```python
inter_set = setA.intersection(setB)
```

```python
print('intersection:',inter_set)
```

```python
# difference
```

```python
print("Difference :", setA.difference(setB))
```

```python
#discard
```

```python
setA.discard(2)
```

```python
print("After discarding A:", setA)
```

```python
#copy
```

```python
copy_set = setA.copy()
```

```python
print("Copied Set:", copy_set)
```

```python
#clear
```

```python
setB.clear()
```

```python
print("After clearing B:", setB)
```
