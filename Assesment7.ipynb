```python
# Task-1
print("1. Numbers from 1 to 50")
for i in range(1, 51):
    print(i)


# Task-2
print("2. Even numbers from 1 to 100")
for i in range(2, 101, 2):
    print(i)


# Task-3
print("3. Sum from 1 to N")
n = int(input("Enter a number: "))
total = 0
for i in range(1, n + 1):
    total += i
print("Sum is:", total)


# Task-4
print("4. Multiplication Table")
num = int(input("Enter number for table: "))
for i in range(1, 11):
    print(num, "x", i, "=", num * i)


# Task-5
print("5. Reverse a String")
text = input("Enter a string: ")
reversed_text = ""
for i in range(len(text) - 1, -1, -1):
    reversed_text += text[i]
print("Reversed string:", reversed_text)


# Task-6
print("6. Count Vowels")
string = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0
for char in string:
    if char in vowels:
        count += 1
print("Number of vowels:", count)


# Task-7
print("7. Pattern of X")
for i in range(1, 6):
    print("*" * i)


# Task-8
print("8. Prime Number Check")
number = int(input("Enter a number: "))
is_prime = True

if number <= 1:
    is_prime = False
else:
    for i in range(2, number):
        if number % i == 0:
            is_prime = False
            break

if is_prime:
    print("It is prime number.")
else:
    print("It is not prime number.")


# Task-9
print("9. Numbers divisible by 3")
numbers = [12, 5, 7, 18, 21, 30]
for i in numbers:
    if i % 3 == 0:
        print(i)


# Task-10
print("10. Break and Continue")
for i in range(1, 21):
    if i % 3 == 0:
        continue
    if i > 15:
        break
    print(i)


# Task-11
print("11. Factorial using while loop")
fact = int(input("Enter number for factorial: "))
factorial = 1

while fact > 0:
    factorial *= num
    fact -= 1

print("Factorial is:", factorial)
```
